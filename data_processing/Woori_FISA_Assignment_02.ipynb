## 3) ë°ì´í„° ì „ì²˜ë¦¬

```python
import pandas as pd

# ğŸ”¹ CSV ê²½ë¡œ
file_path = "/content/drive/MyDrive/fisa_colab_shared_data/combined.csv"

# ğŸ”¹ chunksize ì„¤ì •
chunksize = 1_000_000

# ğŸ”¹ í•„ìš”í•œ ì»¬ëŸ¼
use_columns = [
    "ta_ymd",
    "cty_rgn_no",
    "card_tpbuz_nm_1",
    "card_tpbuz_nm_2",
    "sex",
    "age",
    "amt",
    "cnt"
]

# ğŸ”¹ íƒ€ê²Ÿ ì§€ì—­ì½”ë“œ
target_regions = [41192, 41171, 41210, 41390]

# ğŸ”¹ ì œì™¸ ì—…ì¢…
exclude_categories = ["ì˜ë£Œ/ê±´ê°•", "ê³µê³µ/ê¸°ì—…/ë‹¨ì²´"]

# ğŸ”¹ ì‹œêµ°êµ¬ì½”ë“œ â†’ ì‹œêµ°êµ¬ëª… ë§¤í•‘ dict
region_mapping = {
    41192: "ë¶€ì²œì‹œ",
    41171: "ì•ˆì–‘ì‹œ",
    41210: "ê´‘ëª…ì‹œ",
    41390: "ì‹œí¥ì‹œ"
}

# ğŸ”¹ chunks ìƒì„±
chunks = pd.read_csv(
    file_path,
    chunksize=chunksize,
    encoding="utf-8-sig",
    usecols=use_columns
)

# ğŸ”¹ ì²˜ë¦¬ ë¦¬ìŠ¤íŠ¸
dfs = []

for i, chunk in enumerate(chunks):
    print(f"ğŸ‘‰ Chunk {i+1} ë¡œë”© ì¤‘, shape: {chunk.shape}")

    # ì§€ì—­ í•„í„°ë§
    chunk = chunk[chunk["cty_rgn_no"].isin(target_regions)]

    # ì œì™¸ ì—…ì¢… í•„í„°ë§
    chunk = chunk[~chunk["card_tpbuz_nm_1"].isin(exclude_categories)]

    # ë‚ ì§œ ë³€í™˜
    chunk["ta_ymd"] = pd.to_datetime(chunk["ta_ymd"].astype(str)).dt.strftime("%Y-%m-%d")

    # ì¹´í…Œê³ ë¦¬ ë³€í™˜
    chunk["sex"] = chunk["sex"].astype("category")
    chunk["card_tpbuz_nm_1"] = chunk["card_tpbuz_nm_1"].astype("category")
    chunk["card_tpbuz_nm_2"] = chunk["card_tpbuz_nm_2"].astype("category")

    # ì—°ë ¹ëŒ€ ë³€í™˜
    chunk["age"] = chunk["age"].apply(lambda x: int(x)*10)

    # ì‹œêµ°êµ¬ëª… ì»¬ëŸ¼ ìƒì„±
    chunk["city_name"] = chunk["cty_rgn_no"].map(region_mapping)

    # ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    dfs.append(chunk)

print("âœ… chunk ì²˜ë¦¬ ì™„ë£Œ, ê²°í•© ì‹œì‘...")

# ğŸ”¹ DataFrame ê²°í•©
final_df = pd.concat(dfs, ignore_index=True)
print("âœ… ê²°í•© ì™„ë£Œ:", final_df.shape)

# ğŸ”¹ ì»¬ëŸ¼ëª… ë³€ê²½
final_df = final_df.rename(columns={
    "ta_ymd": "tr_date",
    "card_tpbuz_nm_1": "category_main",
    "card_tpbuz_nm_2": "category_sub",
    "sex": "gender",
    "age": "age_group",
    "amt": "sales_amount",
    "cnt": "sales_count"
})

# ğŸ”¹ city_name ì»¬ëŸ¼ì„ ì›í•˜ëŠ” ìˆœì„œì— ë§ê²Œ ì•ìœ¼ë¡œ ë°°ì¹˜
final_df = final_df[
    [
        "tr_date",
        "city_name",
        "category_main",
        "category_sub",
        "gender",
        "age_group",
        "sales_amount",
        "sales_count"
    ]
]

# ğŸ”¹ ê²°ê³¼ í™•ì¸
print(final_df.head())

# ğŸ”¹ ì €ì¥
output_path = "/content/drive/MyDrive/fisa_colab_shared_data/combined_preprocessing_final.csv"
final_df.to_csv(output_path, index=False, encoding="utf-8-sig")
print("âœ… CSV ì €ì¥ ì™„ë£Œ:", output_path)
```
