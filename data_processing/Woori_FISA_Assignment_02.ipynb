## 3) 데이터 전처리

```python
import pandas as pd

# 🔹 CSV 경로
file_path = "/content/drive/MyDrive/fisa_colab_shared_data/combined.csv"

# 🔹 chunksize 설정
chunksize = 1_000_000

# 🔹 필요한 컬럼
use_columns = [
    "ta_ymd",
    "cty_rgn_no",
    "card_tpbuz_nm_1",
    "card_tpbuz_nm_2",
    "sex",
    "age",
    "amt",
    "cnt"
]

# 🔹 타겟 지역코드
target_regions = [41192, 41171, 41210, 41390]

# 🔹 제외 업종
exclude_categories = ["의료/건강", "공공/기업/단체"]

# 🔹 시군구코드 → 시군구명 매핑 dict
region_mapping = {
    41192: "부천시",
    41171: "안양시",
    41210: "광명시",
    41390: "시흥시"
}

# 🔹 chunks 생성
chunks = pd.read_csv(
    file_path,
    chunksize=chunksize,
    encoding="utf-8-sig",
    usecols=use_columns
)

# 🔹 처리 리스트
dfs = []

for i, chunk in enumerate(chunks):
    print(f"👉 Chunk {i+1} 로딩 중, shape: {chunk.shape}")

    # 지역 필터링
    chunk = chunk[chunk["cty_rgn_no"].isin(target_regions)]

    # 제외 업종 필터링
    chunk = chunk[~chunk["card_tpbuz_nm_1"].isin(exclude_categories)]

    # 날짜 변환
    chunk["ta_ymd"] = pd.to_datetime(chunk["ta_ymd"].astype(str)).dt.strftime("%Y-%m-%d")

    # 카테고리 변환
    chunk["sex"] = chunk["sex"].astype("category")
    chunk["card_tpbuz_nm_1"] = chunk["card_tpbuz_nm_1"].astype("category")
    chunk["card_tpbuz_nm_2"] = chunk["card_tpbuz_nm_2"].astype("category")

    # 연령대 변환
    chunk["age"] = chunk["age"].apply(lambda x: int(x)*10)

    # 시군구명 컬럼 생성
    chunk["city_name"] = chunk["cty_rgn_no"].map(region_mapping)

    # 리스트에 추가
    dfs.append(chunk)

print("✅ chunk 처리 완료, 결합 시작...")

# 🔹 DataFrame 결합
final_df = pd.concat(dfs, ignore_index=True)
print("✅ 결합 완료:", final_df.shape)

# 🔹 컬럼명 변경
final_df = final_df.rename(columns={
    "ta_ymd": "tr_date",
    "card_tpbuz_nm_1": "category_main",
    "card_tpbuz_nm_2": "category_sub",
    "sex": "gender",
    "age": "age_group",
    "amt": "sales_amount",
    "cnt": "sales_count"
})

# 🔹 city_name 컬럼을 원하는 순서에 맞게 앞으로 배치
final_df = final_df[
    [
        "tr_date",
        "city_name",
        "category_main",
        "category_sub",
        "gender",
        "age_group",
        "sales_amount",
        "sales_count"
    ]
]

# 🔹 결과 확인
print(final_df.head())

# 🔹 저장
output_path = "/content/drive/MyDrive/fisa_colab_shared_data/combined_preprocessing_final.csv"
final_df.to_csv(output_path, index=False, encoding="utf-8-sig")
print("✅ CSV 저장 완료:", output_path)
```

# 2. 데이터 분석

## 1) 업종 카테고리 확인

```python
import pandas as pd

# 🔹 전처리된 CSV 경로
file_path = "/content/drive/MyDrive/fisa_colab_shared_data/combined_preprocessing_final.csv"

# 🔹 필요한 컬럼만 불러오기 (메모리 절약)
cols = ["category_main", "category_sub"]

# 🔹 데이터 일부만 불러와서 카테고리 확인 (필요 시 전체로 바꿔도 됨)
df = pd.read_csv(file_path, usecols=cols)

# 🔹 업종 대분류 고유값
print("✅ [대분류 카테고리 목록]")
print(df["category_main"].unique())
print()

# 🔹 업종 소분류 고유값
print("✅ [소분류 카테고리 목록]")
print(df["category_sub"].unique())
print()

# 🔹 대분류 + 소분류 조합 고유값
print("✅ [대분류 + 소분류 조합]")
print(df.drop_duplicates().sort_values(by=["category_main", "category_sub"]).reset_index(drop=True))
```
